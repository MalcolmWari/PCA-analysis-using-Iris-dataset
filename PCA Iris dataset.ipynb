```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("C:/Users/Malco/Desktop/8836201-6f9306ad21398ea43cba4f7d537619d0e07d5ae3/Iris.csv")
df.head(5)
```

```python
#Here i create a variable "X" which is basically me saying to selects all my 4 features of the dataset
#the .values turns it into a numpy array as the input to the PCA algorithm can be more computationally efficient than using pandas objects, because numpy arrays are optimized for numerical computations
X = df.iloc[:, 0:4].values
```

```python
#Here we standardise scales for the features to have a mean of 0 and a standard deviation of 1
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_scaled
```

```python
#I've set my number of components to 3 and also asked to see how much variance is in my data
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
total_var = pca.explained_variance_ratio_.sum() * 100
```

```python
total_var
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Variance (%)') 
plt.show()
```
